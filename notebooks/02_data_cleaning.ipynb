# Split into historical matches and WC2026 fixtures

Since the date which we have is updated and few of the scores are already available, we need split the dateset using the date column. 

```python

historical_matches = df_games[df_games['date'] <'2026-06-11'].copy()
wc2026_fixtures = df_games[df_games['date'] >= '2026-06-11'].copy()
```

# Save to CSV

We shall save these date to csv.

```python

historical_matches.to_csv('../artifacts/interim_data/historical_matches.csv', index=False)
wc2026_fixtures.to_csv('../artifacts/interim_data/wc2026_fixtures.csv', index=False)

print(f"Historical matches: {len(historical_matches)}")
print(f"WC2026 fixtures: {len(wc2026_fixtures)}")
```
